```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

```python
data_path = "../../data/clean_data/clean_"

orderlines = pd.read_csv(f"{data_path}orderlines.csv")
orders = pd.read_csv(f"{data_path}orders.csv")
products = pd.read_csv(f"{data_path}products.csv")
```

```python
discs = (
orderlines
    .merge(products, how="left", on="sku")
    .assign(    disc_amount = lambda x: x["product_price"] - x["ol__unit_price"],
                disc_percent = lambda x: (x["ol_unit_price"] / x["product_price"] * 100))
    .merge(orders, how="left", on="order_id")
)
discounts.head()
```
